```python
import pandas as pd
import sasoptpy as so
import numpy as np
import random
import string
import scipy.stats as ss
import os
from subprocess import Popen
```

```python
data = pd.read_csv("data/gk_season.csv")
```

```python
data = data[data['avg_mins'] > 59].copy()
data
```

```python
def convert_ga_to_xp(ga): # converts poisson rate to expected points
    rv = ss.poisson(mu=ga)
    cs = rv.pmf(0)
    cs2 = rv.pmf(2)
    cs4 = rv.pmf(4)
    cs6 = rv.pmf(6)
    return 2 + 4 * cs - 1 * cs2 - 2 * cs4 - 3 * cs6
```

```python
for gw in range(8,39):
    data[f'GW{gw}_Pts_Est'] = data[f'GW{gw}_GA'].apply(convert_ga_to_xp)
data
```

```python
current_gw = 8
player_ids = data['id'].tolist()
id_dict = data.set_index('id').to_dict(orient='index')
```

```python
def get_random_id(n):
    return ''.join(random.choice(string.ascii_letters + string.digits) for _ in range(n))

def read_solution(m, sol_file):
    with open(sol_file, 'r') as f:
        for v in m.get_variables():
            v.set_value(0)
        for line in f:
            if 'objective value' in line:
                continue
            words = line.split()
            v = m.get_variable(words[1])
            v.set_value(float(words[2]))


def print_solution(m, gws, xp):
    squad = m.get_variable('squad')
    lineup = m.get_variable('lineup')
    # Print solution
    selected_players = []
    gameweek_picks = []
    for p in player_ids:
        entry = {'player': p}
        if squad[p].get_value() > 0.5:
            selected_players.append(p)
            for g in gws:
                entry.update({g: round(lineup[p,g].get_value() * xp[g][p], 3) })
            gameweek_picks.append(entry)
    return selected_players


def solve_N_pick_K_pair_problem(players, measure, budget=10, N=2, K=1, max_iter=1, first_gw=1, last_gw=38, exclude=[], ):
    problem_name = get_random_id(10)
    m = so.Model(name=f'pick_{K}_choose_{N}_{problem_name}')
    gameweeks = list(range(first_gw, last_gw+1))
    # player_ids = players['ID'].tolist()
    squad = m.add_variables(player_ids, vartype=so.binary, name='squad')
    lineup = m.add_variables(player_ids, gameweeks, vartype=so.binary, name='lineup')
    price = players.set_index('id')['price'].to_dict()

    xp = {}
    indexed_players = players.set_index('id')
    for g in gameweeks:
        if isinstance(measure, list):
            if g < first_gw + 9:
                xp[g] = indexed_players[measure[0].format(g)]
            else:
                xp[g] = indexed_players[measure[1].format(g)]
        else:
            xp[g] = indexed_players[measure.format(g)]

    if len(exclude) > 0:
        m.add_constraints((squad[p] == 0 for p in exclude), name='ban_player')

    m.add_constraint(so.expr_sum(squad[p] for p in player_ids) == N, name='select_N')
    m.add_constraints((so.expr_sum(lineup[p, g] for p in player_ids) == K for g in gameweeks), name='pick_K')
    m.add_constraints((lineup[p,g] <= squad[p] for p in player_ids for g in gameweeks), name='valid_picks_only')
    m.add_constraint(so.expr_sum(squad[p] * price[p] for p in player_ids) <= budget, name='budget_con')

    m.set_objective(-so.expr_sum(xp[g][p] * lineup[p,g] for p in player_ids for g in gameweeks), sense='N', name='total_xp')

    m.export_mps(f"tmp/{problem_name}.mps")
    command = f"cbc tmp/{problem_name}.mps solve solu tmp/{problem_name}.sol"
    Popen(command).wait()
    read_solution(m, f'tmp/{problem_name}.sol')
    selected_players = print_solution(m, gameweeks, xp)

    # print(selected_players)

    os.unlink(f'tmp/{problem_name}.mps')
    os.unlink(f'tmp/{problem_name}.sol')

    for it in range(1, max_iter):
        c = m.add_constraint(so.expr_sum(squad[p] for p in selected_players) <= N-1, name=f'cutoff_{it}')
        m.export_mps("{problem_name}.mps")
        Popen(command).wait()
        read_solution(m, f'tmp/{problem_name}.sol')
        selected_players = print_solution(m, gameweeks, xp)

        os.unlink(f'tmp/{problem_name}.mps')
        os.unlink(f'tmp/{problem_name}.sol')

    return {'players': selected_players, 'xp': -1 * m.get_objective_value()}
```

```python
res = solve_N_pick_K_pair_problem(data, measure='GW{}_Pts_Est', budget=10, N=2, K=1, max_iter=1, first_gw=8, last_gw=10)
for r in res['players']:
    print(r)
    print(id_dict[r])
res = solve_N_pick_K_pair_problem(data, measure='{}_Pts', budget=10, N=2, K=1, max_iter=1, first_gw=8, last_gw=10)
for r in res['players']:
    print(r)
    print(id_dict[r])
```
